1. Importando as bibliotecas:

```python
import requests
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize

# Colunas e linhas ilimitadas:
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

2. Setando a URL:

```python
# Obter licitações - Relação das Licitações com os respectivos Licitantes para ANOMODALIDADE == 2019 e objeto contém ALIMENT:

obter_lic = 'https://sistemas.tce.pe.gov.br:443/DadosAbertos/LicitacoesDetalhes!json?ANOMODALIDADE=2019&ESPECIFICACAOOBJETO=ALIMENT'

# Obter licitações - Relação das Licitações para ANOMODALIDADE == 2019 e objeto contém MERENDA:

obter_lic2 = 'https://sistemas.tce.pe.gov.br:443/DadosAbertos/LicitacoesDetalhes!json?ANOMODALIDADE=2019&ESPECIFICACAOOBJETO=MERENDA'
```

```python
# Pegando os dados para ANOMODALIDADE == 2019 e objeto contém ALIMENT (lista1):
r = requests.get(obter_lic2).json()

#Pegando o grupo de dados que realmente interessa:
b = r.get('resposta').get('conteudo')
```

```python
#Montando um dataframe com o resultado da b:
c = b[1].keys()
df_int = pd.DataFrame(b, columns = c)
```

```python
#Verificando:
df_int.head(3)
```

```python
df_int.info()
```

```python
#Olhando nomes únicos de coluna para não me perder:
pd.unique(df_int['NOMEMODALIDADE'])
```

```python
#olhando dispensas:

df_disp = df_int.query('NOMEMODALIDADE == "Dispensa" | NOMEMODALIDADE == "Lei 13.303/2016 - Dispensa"')
df_disp.head(3)
```

```python
#Puxando os dados do contrato:
url = 'https://sistemas.tce.pe.gov.br/DadosAbertos/Contratos!json?AnoContrato=2019&NumeroProcesso=11&CodigoUG=605'

r = requests.get(url).json()
b = r.get('resposta').get('conteudo')
c = b[1].keys()
```

```python
#Criando o dataframe:

df_con = pd.DataFrame(b, columns = c)
df_con['AnoProcesso'] = df_con['AnoProcesso'].astype(str).astype(int)
df_con = df_con.query('AnoProcesso == 2019')
```

```python
# Verificando:

df_con.head(1)
```
